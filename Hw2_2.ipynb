```python
#12.Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
dfg = python_data.groupby(python_data['Должность'], group_keys = True)
```

```python
#13.Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg
# і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
dfnew =dfg['Зарплата.в.месяц'].agg(['min', 'max'])
print(dfnew)
```

```python
#14.Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(i):
    return i.mean()
dfnew['avg'] = dfnew.apply(fill_avg_salary, axis=1)
print(dfnew)
```

```python
#15.Створіть описову статистику за допомогою методу describe для нового стовпчика.
print(dfnew['avg'].describe())
```
