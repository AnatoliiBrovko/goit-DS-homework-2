```python
import pandas as pd
df = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", match="Коефіцієнт народжуваності в регіонах України")
df = df[0]
print(df)
```

```python
#1.Вивести перші рядки таблиці за допомогою методу head
print(df.head())
```

```python
#2.Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
print(df.shape)
```

```python
#3.Замініть у таблиці значення "—" на значення NaN
import numpy as np
dfn = df.replace(to_replace = '—' , value = np.NaN)
print(dfn)
```

```python
#4.Визначте типи всіх стовпців за допомогою dataframe.dtypes
print(dfn.dtypes)
```

```python
#5.Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
dfnn = dfn.apply(pd.to_numeric, errors='ignore')
print(pd.DataFrame(dfn).dtypes)
print(pd.DataFrame(dfnn).dtypes)
```

```python
#6.Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
print((dfn.isnull().sum())/len(dfn) * 100)
```

```python
#7.Видаліть з таблиці дані по всій країні, останній рядок таблиці
empl = dfnn.drop([27], axis=0)
print(empl)
```

```python
#8.Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
dff = empl.fillna(empl.mean ()).round(1)
print(dff)
```

```python
#9.Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
m = dff["2019"].mean()
print(m)
dfm = dff[dff["2019"] > m]
print(dfm["Регіон"])
```

```python
#10.У якому регіоні була найвища народжуваність у 2014 році?
max = dff["2014"].max()
print(max)
dfm = dff[dff["2014"] == max]
print(dfm["Регіон"])
```

```python
#11.Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
import matplotlib.pyplot as plt
plt.barh(dff["Регіон"], dff["2019"])
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(dff['Регіон'], dff['1950'], label='1950', color = 'aqua')
plt.plot(dff['Регіон'], dff['1960'], label='1960', color = 'forestgreen')
plt.plot(dff['Регіон'], dff['1970'], label='1970', color = 'pink')
plt.plot(dff['Регіон'], dff['1990'], label='1990', color = 'orangered')
plt.plot(dff['Регіон'], dff['2000'], label='2000', color = 'purple')
plt.plot(dff['Регіон'], dff['2012'], label='2012', color = 'royalblue')
plt.plot(dff['Регіон'], dff['2014'], label='2014', color = 'gold')
plt.plot(dff['Регіон'], dff['2019'], label='2019', color = 'lime')
plt.title('Рівень народжуваності по регіонах України за роками з 1950-2019')
plt.ylabel('Народжуваність', fontsize='small')
plt.xlabel('Регіон', fontsize='small')
plt.xticks(rotation=90)
plt.grid()
plt.legend()
plt.show()
```

```python
dnp = dff[dff['Регіон'] == 'Дніпропетровська']
plt.plot(dnp.iloc[0,1:])
plt.title('Динаміка народжуваності у Дніпропетровській області')
plt.xlabel('Рік')
plt.ylabel('Кількість народжених')
plt.grid()
plt.show()
```
